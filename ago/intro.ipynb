```python
from sklearn import datasets
import pandas as pd
import numpy as np

digits = datasets.load_digits()
df = pd.DataFrame(np.column_stack([digits['data'], digits['target']]), columns=digits['feature_names'] + ['target'])

df.describe()
```

```python
#clf is classifier estimator
#svm.SVC implements support vector classification
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
#use 1796 to train, leave one to test
clf.fit(digits.data[:-1], digits.target[:-1])

#predict the one
predicted = clf.predict(digits.data[-1:])

predicted.shape
print(predicted)
print(digits.target[-1:])

# very accurate
```
